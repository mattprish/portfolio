```python
import numpy as np
import pandas as pd

from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.impute import SimpleImputer
```

```python
ss = pd.read_csv('/content/sample_submission.csv')
ss.head()
```

```python
df = pd.read_csv('/content/train.csv')
df_test = pd.read_csv('/content/test.csv')
df.head()
```

```python
cols = df.columns.tolist()
cols.remove('PassengerId')
cols.remove('Name')
cols.remove('Transported')

cols.remove('HomePlanet')
cols.remove('Cabin')
cols.remove('Destination')

X = df[cols]
y = df['Transported']

print(X.shape)
X.head()
```

```python
X.isna().sum()
```

```python
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
X = pd.DataFrame(imp_mean.fit_transform(X))
X.columns = cols
print(X.shape)
```

```python
lr = LogisticRegression(max_iter=300, C=100).fit(X, y)
```

```python
y_pred = lr.predict(X)
acc = accuracy_score(y_pred, y)
print(acc)
np.unique(y_pred, return_counts=True)
```

```python
id = df_test['PassengerId']
df_test = df_test[cols]
df_test = pd.DataFrame(imp_mean.transform(df_test))

df_test.columns = cols
# df_test.head()
```

```python
y_pred = lr.predict(df_test)
```

```python
np.unique(y_pred, return_counts=True)
```

```python
df_submit = pd.DataFrame({'1': id, '2': y_pred})
df_submit.columns = ['PassengerId', 'Transported']
df_submit.head()
```
